```python
from numpy.lib.function_base import percentile
def remove_outliers(No_of_Dependents):
  sorted(No_of_Dependents)
  Q1, Q3 = np.percentile(No_of_Dependents, [25, 75])
  iqr = Q3 - Q1
  lower_range = Q1 - (1.5*iqr)
  return lower_range
```

```python
ir = remove_outliers(austo_mc['No_of_Dependents'])
print ('lower range', ir)
```

```python
austo_mc['No_of_Dependents'] = np.where(austo_mc['No_of_Dependents']<ir, ir, austo_mc['No_of_Dependents'])
```

```python
def remove_outliers(Total_salary):
  sorted(Total_salary)
  Q1, Q3 = np.percentile(Total_salary, [25, 75])
  iqr = Q3 - Q1
  upper_range = Q3 + (1.5*iqr)
  return upper_range
```

```python
ur = remove_outliers(austo_mc['Total_salary'])
print ('upper range', ur)
```

```python
austo_mc['Total_salary'] = np.where(austo_mc['Total_salary']>ur, ur, austo_mc['Total_salary'])
```
